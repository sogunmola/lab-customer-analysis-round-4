```python
# Library Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from scipy import stats
```

```python
# File Upload
Mktcus = pd.read_csv("marketing_customer_analysis.csv")
```

## Data Cleaning

```python
# Header Check
Mktcus.head()
```

```python
# Standardizing headers - Removing spaces
Mktcus.columns = Mktcus.columns.str.replace(' ','')
```

```python
# Standardizing headers - Making all lower case
Mktcus.columns = Mktcus.columns.str.lower()
```

```python
# Header check
Mktcus.head()
```

## Analysis

```python
#1/ Get the numeric data (Created df "NUM" to store numeric data)
NUM = Mktcus.select_dtypes('number')
print("Numeric columns ", list(NUM.columns))
```

```python
#2/ Check for normality by plotting on histograms
NUM.hist()
plt.show()
```

```python
#3.1/ Check multicollinearity (Pearsons)
pears_matrix = NUM.corr(method='pearson')
fig, ax = plt.subplots(figsize = (10,8))
ax = sns.heatmap(pears_matrix, annot=False)
plt.show()
```

```python
#3.2/ Check multicollinearity (Spearman)
spear_matrix = NUM.corr(method='spearman')
fig, ax = plt.subplots(figsize = (10,8))
ax = sns.heatmap(spear_matrix, annot=False)
plt.show()
```

```python
#3.2/ Check multicollinearity (Kendall)
kend_matrix = NUM.corr(method='kendall')
fig, ax = plt.subplots(figsize = (10,8))
ax = sns.heatmap(kend_matrix, annot=False)
plt.show()
```
